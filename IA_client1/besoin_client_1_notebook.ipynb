```python
######## import des bibliothèques ########

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import normalized_mutual_info_score
from sklearn import model_selection
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn import model_selection
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import davies_bouldin_score
```

```python
######## lecture et analyse du dataset ########

dataset = pd.read_csv('Données_V5.csv', sep=';')

print(type(dataset))
print(dataset.head())
print(dataset.info())
print(dataset.shape)
print(dataset.info)
```

```python
######## modèle k-means #########

#on ne garde que la colonne avec les hauteurs des arbres pour les classer par taille
haut_col = [dataset.columns[8]]

#on attribut à x la colonne qu'on a extrait du dataset
x = dataset[haut_col]

#on retire les NaN car cela peut créer des erreurs par la suite
x = x.dropna()

#on standardise et normalise les données pour que la moyenne/variance reste cohérente.
#(exemple : si toutes les valeurs sont entre 0 et 10 et qu'il y a une valeur à 2000, tout les calculs seront biaisés.)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)

#on initialise les variables pour le calcul des meilleurs paramètres.
#le silouhette score doit être être proche de 1 donc on initialise à 0.
#le Calinski-Harabasz doit être le plus grand possible donc on l'initialise à 0.
#le Davies-Bouldin doit être proche de 0 sans valeur maximale donc nous l'initialisons à None.
#on test 4 tailles de clusters entre 2 et 5 donc on initialsie le pool de test.
#on initialise des variable best_k pour déterminer quel est le meilleur cluster par rapport à son score.
max_sil = 0
Ks_test = [2,3,4,5]
best_k = 0
max_kav = 0
best_k_kav = 0
max_dav = None
best_k_dav = 0

#on créer une boucle qui va prendre en valeur k parmi le pool initié ks_test pour déterminer mes kmeans par nomber de cluster
for k in Ks_test:



#on fait le kmeans et on enjolive la présentation
  print("##################### k=",k, "##########################")
  kmeans = KMeans(n_clusters=k, random_state=42)
  kmeans.fit(X_scaled)
  klabel = kmeans.fit_predict(X_scaled)

  #on détermine la taille des clusters
  taille_cluster = pd.Series(klabel).value_counts().sort_index().to_dict()
  print(f"Taille des clusters pour k={k}: {taille_cluster}")


  #on affiche pour chaque métrique sa valeur par cluster
  print("la Silhouette score for k=",k,"est :",silhouette_score(X_scaled, klabel))
  print("le calinski score pour k=",k,"est :",calinski_harabasz_score(X_scaled, kmeans.labels_))
  print("le davies score pour k=",k,"est :",davies_bouldin_score(X_scaled, klabel))
  print("\n")

  #on compare cherche à savoir quelle est la meilleuer valeur par métrique
  #on fait des boucles qui vont comparer la valeur de la métrique actuelle à celle considéré comme la meilleure
  #on garde son nombre de cluster et sa valeur
  if max_sil < silhouette_score(X_scaled, klabel):
    max_sil = silhouette_score(X_scaled, klabel)
    best_k = k
  else :
    max_sil = max_sil
    best_k = best_k

  if calinski_harabasz_score(X_scaled, kmeans.labels_) > max_kav:
    max_kav = calinski_harabasz_score(X_scaled, kmeans.labels_)
    best_k_kav = k
  else :
    best_kav = best_kav
  if davies_bouldin_score(X_scaled, klabel) < max_dav:
    max_dav = davies_bouldin_score(X_scaled, klabel)
    best_k_dav = k
  else :
    best_k_dav = best_k_dav

#on affiche le résultat final des meilleures métriques par nombre de cluster.
print( "meilleur silouhette score = ", max_sil, "avec pour nombre de clusters = ", best_k)
print( "meilleur calinski harabasz score = ", max_kav, "avec pour  nombre de clusters = ", best_k_kav)
print( "meilleur davies score = ", max_dav,"avec pour nombre de clusters = ", best_k_dav)
```

```python
############# détermination de la taille des arbres par cluster #############

#on prend le kmeans final pour 2 clusters
final_kmeans2 = KMeans(n_clusters=2, random_state=42)
final_klabel2 = final_kmeans2.fit_predict(X_scaled)

#on détermine la série avec pour index son numéro d'itération
#pour que la série créée corresponde à la taille de l'arber testé dans le dataset
#on créer une colonne 'culster_taille' qui va avoir une valeur binaire en fonction du culster dans lequel l'arbre se trouve
cluster_series_k2 = pd.Series(final_klabel2, index=x.index)
dataset['cluster_taille'] = cluster_series_k2

#on créer un mapping pour que la variable "taille cluster2" possède la taille enregistrée de manière lisible
cluster_mapping_k2 = {0: 'grand', 1: 'petit'}
dataset['cluster_taille2'] = dataset['cluster_taille'].map(cluster_mapping_k2)

#on prend le kmeans final pour 3 clusters
final_kmeans3 = KMeans(n_clusters=3, random_state=42)
final_klabel3 = final_kmeans3.fit_predict(X_scaled)

#on détermine la série avec pour index son numéro d'itération
#pour que la série créée corresponde à la taille de l'arber testé dans le dataset
#on créer une colonne 'culster_taille' qui va avoir une valeur binaire en fonction du culster dans lequel l'arbre se trouve
cluster_series_k3 = pd.Series(final_klabel3, index=x.index)
dataset['cluster_taille3'] = cluster_series_k3


#on créer un mapping pour que la variable "taille cluster3" possède la taille enregistrée de manière lisible
cluster_mapping_k3 = {0: 'moyen', 1: 'petit', 2: 'grand'}
dataset['cluster_taille3'] = dataset['cluster_taille3'].map(cluster_mapping_k3)

print("DataFrame avec les nouvelles colonnes de cluster:")
print(dataset[['haut_tot', 'cluster_taille', 'cluster_taille2', 'cluster_taille3']].head(10))

```

```python
#on exporte au format csv le dataset ajusté avec les nouvelles colonnes, 'cluster tialle', 'cluster taille2' et 3
dataset.to_csv("dataset_ajuste.csv", index=False)
```
